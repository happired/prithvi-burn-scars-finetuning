# Visualizing model output

```python
import matplotlib.pyplot as plt
import rasterio
import matplotlib

NO_DATA = -9999
NO_DATA_FLOAT = 0.0001
PERCENTILES = (0.1, 99.9)
```

```python
# reads & returns raster image

def load_raster(path, crop=None):
    with rasterio.open(path) as src:
        img = src.read()

        # load first 6 bands
        img = img[:6]

        img = np.where(img == NO_DATA, NO_DATA_FLOAT, img)
        if crop:
            img = img[:, -crop[0]:, -crop[1]:]
    return np.squeeze(img)
```

```python
# normalizes raster image to RGB

def enhance_raster_for_visualization(raster, ref_img=None):
    if ref_img is None:
        ref_img = raster
    channels = []
    for channel in range(raster.shape[0]):
        valid_mask = np.ones_like(ref_img[channel], dtype=bool)
        valid_mask[ref_img[channel] == NO_DATA_FLOAT] = False
        mins, maxs = np.percentile(ref_img[channel][valid_mask], PERCENTILES)
        normalized_raster = (raster[channel] - mins) / (maxs - mins)
        normalized_raster[~valid_mask] = 0
        clipped = np.clip(normalized_raster, 0, 1)
        channels.append(clipped)
    clipped = np.stack(channels)
    channels_last = np.moveaxis(clipped, 0, -1)[..., :3]
    rgb = channels_last[..., ::-1]
    return rgb
```

```python
#visualize model inference

norm = matplotlib.colors.Normalize(vmin=0, vmax=2)

fig, (ax1, ax2, ax3) = plt.subplots(3, 5, figsize=(20, 10))

modelInferences = [ "subsetted_512x512_HLS.S30.T11TME.2019222.v1.4_merged_pred.tif", "subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged_pred.tif","subsetted_512x512_HLS.S30.T17RMH.2021103.v1.4_merged_pred.tif"]
checkpointModelInferences = ["subsetted_512x512_HLS.S30.T11TME.2019222.v1.4_merged_pred.tif","subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged_pred.tif",  "subsetted_512x512_HLS.S30.T17RMH.2021103.v1.4_merged_pred.tif"]
originalImages = ["subsetted_512x512_HLS.S30.T11TME.2019222.v1.4_merged.tif", "subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4_merged.tif","subsetted_512x512_HLS.S30.T17RMH.2021103.v1.4_merged.tif"]
groundTruths = ["subsetted_512x512_HLS.S30.T11TME.2019222.v1.4.mask.tif", "subsetted_512x512_HLS.S30.T11SMU.2020299.v1.4.mask.tif","subsetted_512x512_HLS.S30.T17RMH.2021103.v1.4.mask.tif"]

counter = 0

for plot in (ax1, ax2, ax3):

    #iterate through images for each category and concatenate strings to create file path

    for subplot in plot:
      subplot.axis('off')

    plot[0].imshow(enhance_raster_for_visualization(load_raster("inference/inference-images/"+originalImages[counter])))
    plot[0].set_title("Original Image") if counter == 0 else None

    plot[1].imshow(load_raster("inference/ground-truths/"+groundTruths[counter]), norm=norm, cmap="turbo")
    plot[1].set_title("Ground Truth") if counter == 0 else None

    plot[2].imshow(load_raster("results/inference-test/"+modelInferences[counter]), norm=norm, cmap="turbo")
    plot[2].set_title("1000-Iteration Model Inference") if counter == 0 else None

    plot[3].imshow(load_raster("results/Prithvi-100M-burn-scar/"+checkpointModelInferences[counter]), norm=norm, cmap="turbo")
    plot[3].set_title("Checkpoint Model Inference") if counter == 0 else None

    plot[4].imshow(enhance_raster_for_visualization(load_raster("inference/inference-images/"+originalImages[counter])))
    plot[4].imshow(load_raster("results/inference-test/"+modelInferences[counter]), norm=norm, cmap="turbo", alpha=0.15)
    plot[4].set_title("Overlay") if counter == 0 else None

    counter += 1

```
